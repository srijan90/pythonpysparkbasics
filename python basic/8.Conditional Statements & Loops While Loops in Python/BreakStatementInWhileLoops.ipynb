The `break` statement provides you with the opportunity to exit out of a loop when an external condition is triggered

The programme here will be a infinite loop

```python
number = 20

while number < 25:
    print(number)
    
```

Now we can come out from the loop using break statement

```python
number = 20

while number < 25:
    print(number)
    break
```

Using range function break out of a whileloop

```python
num = 10

while num in range(10, 100):
    
    print(num)
    
    if num == 50:
        break
        
    num = num + 10
    
print("out of the loop ")
```

In this example, we have a list of vegetable which contains the fruit “Orange”. We examined from the first element to the end, if we find the element we will print the position of the fruit and delete the fruit from the list and escape the loop.<br>Then print the updated vegetable list

```python
clean_haystack =["hay", "hay", "hay", "hay",
                 "hay", "hay", "hay", "hay"]

unclean_haystack =["hay", "hay", "hay", "hay",
                   "hay", "needle", "hay", "hay"]

searching_for = "needle"
```

```python
maximum = len(clean_haystack)

i = 0

while i < maximum:
    
    if searching_for == clean_haystack[i]:
        print("The %s is at index %i:" % (searching_for, i))
        
        del clean_haystack[i]
        break
        
    i = i + 1

else:
    print("\nThe needle was not found :(")
```

```python
maximum = len(unclean_haystack)

i = 0

while i < maximum:
    
    if searching_for == unclean_haystack[i]:
        print("The %s is at index %i :)" % (searching_for, i))
        
        del unclean_haystack[i]
        print ('The updated haystack is:', unclean_haystack)
        break
        
    i = i + 1

else:
    print("\nThe needle was not found :(")
```

Printing all the prime numbers between 2 and the given value using break statement

```python
max_num = int(input('enter an integer: '))
num = 2

print("prime numbers upto %i: "% max_num)

while num < max_num + 1: 
    
    isPrime = True
    count = 2
    
    while count < num:
        if num % count == 0:
            isPrime = False
            break
            
        count += 1

    if isPrime:
        print(num)

    num += 1
```

#### Use of pass in while loop<br>
When an external condition is triggered, the pass statement allows you to handle the condition without the loop being impacted in any way; all of the code will continue to be read unless a break or other statement occurs.

```python
num_value = 10

while num_value > 0:              
    print ("current value is :", num_value)
    
    if num_value == 5:
        pass
        print ('this is a pass block')
        
    num_value = num_value - 1

print ("good bye!")
```
