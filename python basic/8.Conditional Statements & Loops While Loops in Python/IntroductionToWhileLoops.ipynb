#### With the while loop we can execute a set of statements as long as a condition is true.

* Print a statement using while loop
* The Infinite Loop
* Single-line while loop with condition
* Nested While Loops
* python `range` function in while loop

Here the condition is false<br>
So the loop didn't execute the statement

```python
number = 20

while number < 20:
    print(number)
```

Here the condition is always true<br>
So the loop will keep running forever, It will never end , Such a loop is called an infinite loop.

```python
number = 20

while number < 25:
    print(number)
```

Here after printing the number we are increasing the value of the number<br>
So the condition will be false and we come out from the loop

```python
number = 20

while number < 25:
    print(number)
    
    number = 26
```

Here we are changing the value of the number but the condition is still true,<br>
So this also will be a infinite loop

```python
number = 20

while number < 25:
    print(number)
    
    number = 24
```

We can use some arithmetic operations in condition as well

```python
num = 1

while num <= 10:
    
    print(num ** 2)
    
    num += 1
```

Print a name certain number of times

If statements within while loops

```python
num = 0

while num in range(0, 10):
    
    if num % 2 == 0:
        print("even number: ", num)
        
    num = num + 1
```

Check the digits of an integer

```python
value_int = int(input("enter an integer: "))

length = 0

while value_int > 0:
    
    value_int //= 10 
   
    length += 1
    
print("number of digits: " , length) 
```

Getting the sum of the digits of a number

```python
num = int(input("enter a number: "))

total = 0

while (num !=0):
    
    remainder = num % 10
    
    total = total + remainder
    num = num // 10

print("the total sum of digits is: ", total)
```

Using else Statement with whileLoop

```python
count = 0

while count < 8:
    
    print (count, " is  less than 8")
    
    count = count + 1
    
else:
    print (count, " is not less than 8")
```

Like nested if condition<br>
We can write a nested while loop in python as well<br>
Here we are finding odd and even numbers from a certain range

```python
x = 1

while x < 5:
    
    y = 1
    
    while y < 5:
        print(x,y)
        y = y + 1
        
    x = x + 1
```
